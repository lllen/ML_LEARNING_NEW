1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x_symbol = sp.symbols('x')
func_x_symbol = 10 * x_symbol - 5
derivative_func = sp.diff(func_x_symbol, x_symbol)

print('Derivative of 10x - 5: ', derivative_func);


x = np.linspace(0, 10, 10)
func_x = 10 * x - 5
derivative_func_x = 10 * np.ones_like(x)

plt.figure(figsize=(10, 5))
plt.plot(x, func_x, label='func_x', color='blue')
plt.plot(x, derivative_func_x, label='derivative_func_x', color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x - кількість годин")
plt.ylabel("y - оцінка")
plt.legend()
plt.grid(True)
plt.show()
```

**Оскільки похідна 10, бачимо що з кожною додатковою годиною буде зростати і оцінка, тобто швидкість зростання рівномірна і присутня і чим більше годин приділено навчанню тим вищою буде оцінка в результаті.**

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
func = sp.sin(x)

dr_func = sp.diff(func, x)

print('Функція:', func)
print('Похідна:', dr_func)

func_num = sp.lambdify(x, func, 'numpy')
dr_func_num = sp.lambdify(x, dr_func, 'numpy')
x_vals = np.linspace(0, 10, 100)
y_vals = func_num(x_vals)
dy_vals = dr_func_num(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = cos(x)", color='green')

plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = sp.symbols('x')
func = sp.ln(x)

dr_func = sp.diff(func, x)

print('Функція:', func)
print('Похідна:', dr_func)

func_num = sp.lambdify(x, func, 'numpy')
dr_func_num = sp.lambdify(x, dr_func, 'numpy')
x_vals = np.linspace(0, 10, 100)
y_vals = func_num(x_vals)
dy_vals = dr_func_num(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x)", color='green')

plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
func = x**2 - 5*x + 1

dr_func = sp.diff(func, x)

print('Функція:', func)
print('Похідна:', dr_func)

func_num = sp.lambdify(x, func, 'numpy')
dr_func_num = sp.lambdify(x, dr_func, 'numpy')
x_vals = np.linspace(0, 100, 100)
y_vals = func_num(x_vals)
dy_vals = dr_func_num(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x)", color='green')

plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x - кількість вироблених одиниць")
plt.ylabel("y - витрати компанії")
plt.legend()
plt.grid(True)
plt.show()
```
